## Disaster table cleaning

```python
pd.get_dummies(dis_df.disastertype)
```

```python
dis_extended = pd.get_dummies(dis_df, columns=['disastertype'])
```

```python
disasters = dis_extended.drop(['adm1', 'adm2', 'adm3', 'geolocation', 'level', 'historical', 'hist_country', 'iso3', 'gwno', 'geo_id', 'location', 'disasterno', 'latitude', 'longitude'], axis=1)
```

```python
disasters.set_index('id', drop=True, inplace=True)
disasters
```

```python
disasters.drop_duplicates(inplace=True)
```

```python
disasters.sort_values(by=['year', 'country'], inplace=True)
disasters
```

```python
dis_used = disasters.drop('country', axis=1)
dis_used = dis_used[dis_used['year'] <= 2015]
dis_used
```

```python
dis_used = dis_used.groupby(by='year').sum()
dis_used
```

## Temperature data cleaning

```python
temps_df.info()
```

```python
temps_df.head()
```

```python
temperatures = temps_df.set_index('dt', drop=True)
temperatures
```

```python
temp_used = temperatures['1960-01-01':]
temp_used
```

```python
years = temp_used.index.str.split('-').values.tolist()
temp_date = pd.DataFrame(years, columns = ['year', 'month', 'day'])
temp_date.drop(['day'], axis=1, inplace=True)
temp_date
```

```python
temp_used = temp_used.reset_index()
temp_used.drop(['dt'], axis=1, inplace=True)
temp_used
```

```python
frames = [temp_date, temp_used]
temps = pd.concat(frames, axis=1)
temps
```

```python
temps = temps.groupby(by='year').mean()
temps
```

## Combine 2 data sets

```python
temps
```

```python
dis_used
```

```python
dis_used.reset_index(inplace=True)
dis_used.head()
```

```python
temps.reset_index(inplace=True)
temps.head()
```

```python
temps.year=temps.year.astype(int)
```

```python
global_data = dis_used.merge(temps, on='year')
global_data
```

```python
global_data.to_csv('global_disaster_vs_temps.csv', index=False)
```
